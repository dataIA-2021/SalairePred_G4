```python
def salary_stripper(dataframe, column):
    dataframe[str(column)] = dataframe[str(column)].replace({'\€':''}, regex = True)
    dataframe[str(column)].replace(regex=True,inplace=True,to_replace=r'\D',value=r' ')
    dataframe[str(column)] = dataframe[str(column)].str.replace(' ',',')
    dataframe = dataframe.join(dataframe[str(column)].str.split(',,,', 1, expand=True).rename(columns={0:'Low', 1:'High'}))
    dataframe['Low'] = dataframe['Low'].str.replace(',','')
    dataframe['Low'] = dataframe['Low'].astype('float64')
    dataframe.drop(str(column), axis=1, inplace=True)
    dataframe['High'] = dataframe['High'].str.replace(',','')
    dataframe['High'] = dataframe['High'].apply(pd.to_numeric)
    dataframe['Salaire_annuel_moyen'] = dataframe[['Low', 'High']].mean(axis=1)
    dataframe = dataframe.drop(columns=['Rating','Low','High'])
    return dataframe
```

```python
class Preprocess():
    def __init__(self,df):
        
        self.df_ = df   
        
        def isFreelance(txt):
            if bool(re.search("freelance", txt)):
                return 1
            else:
                return 0

        def isDataAnalyst(txt):
            if bool(re.search("(?=.*scientist)(?=.*data)", txt)):
                return 1
            else:
                return 0

        def isDataScintist(txt):
            if bool(re.search("(?=.*analyst)(?=.*donné)", txt))|bool(re.search("(?=.*analyst)(?=.*data)", txt)):
                return 1
            else:
                return 0

        def isConfirm(txt):
            if bool(re.search("(?=.*confirm)", txt))|bool(re.search("(?=.*senior)", txt)):
                return 1
            else:
                return 0
        
        self.df_['LocNum'] = df['Location'].str.extract(r'(\d{2})', expand=False)
        # lb_encod = OrdinalEncoder()
        # self.df_['LocNum'] = lb_encod.fit_transform(self.df_['LocNum'])
        
        self.df_['freelance'] = self.df_['Title'].map(isFreelance)
        self.df_['DataScientist'] = self.df_['Title'].map(isDataAnalyst)
        self.df_['DataAnalyste'] = self.df_['Title'].map(isDataScintist)
        self.df_['senior'] = self.df_['Title'].map(isConfirm)
        self.df_.drop(self.df_[self.df_.Average<20000].index, inplace=True)
        
        self.X = self.df_.drop(columns=['Average','Title','Location','Synopsis','Company','Rating','Low','High'])
        self.y = self.df_['Average']
    
    
    def getDf(self):
        return self.df_
        
    def getFeatures(self):
        return self.X
    
    def getTarget(self):
        return self.y
```

```python
preproc = Preprocess(df)
df = preproc.getDf()
X = preproc.getFeatures()
y = preproc.getTarget()
```
